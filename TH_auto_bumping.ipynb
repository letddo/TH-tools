```python
userid = "ID here" #@param {type:"string"}

userpw = "PW here" #@param {type:"string"}

folder = "Folder URL" #@param {type:"string"}

# install chromium, its driver, and selenium
!apt update &> /dev/null
!apt install chromium-chromedriver &> /dev/null
!pip install selenium &> /dev/null
# set options to be headless, ..
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import getpass
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

print("setting complete")


# step 1 -----------------------------------------------------
# open it, go to a website, and get results
driver = webdriver.Chrome(options=options)
driver.get("https://toyhou.se/")
driver.find_element(By.XPATH, '//*[@id="container"]/div/div/div[3]/div[2]/a').click()

time.sleep(0.5)
driver.find_element(By.ID, 'username').send_keys(userid)
driver.find_element(By.ID, 'password').send_keys(userpw)
driver.find_element(By.XPATH, '//*[@id="modal-login"]/div/div/div[2]/form/div[3]/input[2]').click()

#print(driver.page_source)  # results
# divs = wd.find_elements_by_css_selector('div')




# step 2 -----------------------------------------------------
def getPage(link):
    bumplink = link[18:]
    tmp = bumplink.find('.')
    bumplink = bumplink[0:tmp]

    driver.get('https://toyhou.se/~characters/trade-listing/' + bumplink)
    time.sleep(0.5)


def unlist():
    # 1 2 3 4
    checklist = list()
    tmp = driver.find_elements(By.CLASS_NAME, 'row')

    tmplist = tmp[1].find_elements(By.CLASS_NAME, 'col-lg-4')
    for i in tmplist:
        checklist.append(i.find_element(By.CLASS_NAME, 'type-checkbox').get_attribute('checked'))

    tmplist = tmp[2].find_elements(By.CLASS_NAME, 'col-lg-2')
    for i in tmplist:
        checklist.append(i.find_element(By.CLASS_NAME, 'form-control').get_attribute('value'))

    tmplist = tmp[3].find_elements(By.CLASS_NAME, 'col-lg-2')
    for i in tmplist:
        checklist.append(i.find_element(By.TAG_NAME, 'input').get_attribute('checked'))

    tmplist = tmp[4].find_elements(By.CLASS_NAME, 'col-lg-2')
    for i in tmplist:
        checklist.append(i.find_element(By.TAG_NAME, 'input').get_attribute('checked'))

    try:
        driver.find_element(By.XPATH, '//*[@id="content"]/form[2]/div/input').click()
    except:
        return 0;

    #print(checklist)

    return checklist

def relist():
    number = 0

    tmp = driver.find_elements(By.CLASS_NAME, 'row')

    tmplist = tmp[1].find_elements(By.CLASS_NAME, 'col-lg-4')
    for i in tmplist:
        if (checklist[number] != None):
            i.find_element(By.CLASS_NAME, 'type-checkbox').click()
        number += 1

    tmplist = tmp[2].find_elements(By.CLASS_NAME, 'col-lg-2')
    for i in tmplist:
        if (checklist[number] != None):
            i.find_element(By.CLASS_NAME, 'form-control').send_keys(checklist[number])
            number += 1


    tmplist = tmp[3].find_elements(By.CLASS_NAME, 'col-lg-2')
    for i in tmplist:
        if (checklist[number] != None):
            i.find_element(By.TAG_NAME, 'input').click()
        number += 1


    tmplist = tmp[4].find_elements(By.CLASS_NAME, 'col-lg-2')
    for i in tmplist:
        if (checklist[number] != None):
            i.find_element(By.TAG_NAME, 'input').click()
        number += 1


    driver.find_element(By.XPATH, '//*[@id="submit"]').click()

# ----------------------------------------------

bumplist = list()
num = 1
flag = 0

while (1):
  if flag == 1: break
  driver.get(folder+"?page="+str(num))
  num += 1
  bumpfolder = driver.find_elements(By.CLASS_NAME, 'thumb-character-name')
  #print(num);
  if (bumpfolder == []): break
  for i in bumpfolder:
      aTag = i.find_element(By.TAG_NAME, 'a')
      if aTag.get_attribute('href') in bumplist:
        flag = 1;
        break
      bumplist.append(aTag.get_attribute('href'))

for i in bumplist:
    print(i)
    getPage(i)
    checklist = unlist()
    #print(checklist)
    if checklist != 0:
        relist()

print("Done")
```
